```python
import pandas as pd
data = pd.read_csv('graduate_employability.csv')
print(data.head())
print(data.describe())
```

```python
print(f"\n--Dataset Analysis--")
rows, columns = data.shape
print(f'The dataset contains {rows} rows and {columns} columns.')
print(data.isnull().sum())
```

```python
# Distribution by education level
level_dist = data['Education Level'].value_counts()
print("\nEducation level distribution:")
print(level_dist)

# Employability by education level
employability_by_level = data.groupby('Education Level')['GM (%)'].mean()
print("\nAverage employability by education level:")
print(employability_by_level)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns   
#visualize the eduaction level distribution
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='Education Level', order=level_dist.index)
plt.title('Distribution of Education Levels')
plt.xlabel('Education Level')
plt.ylabel('Count') 
plt.show()

#Visualize using pie chart
plt.figure(figsize=(8, 8))
plt.pie(level_dist, labels=level_dist.index, autopct='%1.1f%%', startangle=140)
plt.title('Education Level Distribution')
plt.show()
```

```python
# Graduate employability rate distribution based in Malaysia 
# Top institutions with most courses
top_institutions = data['Institutions'].value_counts().head(20)
print("Top 20 institutions by number of courses:")
print(top_institutions)

# Institutions with 100% employability
institutions_100 = data[data['GM (%)'] == 100]['Institutions'].unique()
print(f"Number of institutions with 100% employability: {len(institutions_100)}")
```

```python
# Visualize the distribution of employability rates
plt.figure(figsize=(10, 6))
sns.boxplot(x='Education Level', y='GM (%)', data=data)
plt.title('Employability Rates by Education Level') 
plt.xlabel('Education Level')
plt.ylabel('Graduate Employability Rate (%)')
```

```python
sns.pairplot(data, hue='Education Level', vars=['GM (%)'])
```
