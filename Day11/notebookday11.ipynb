```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
#Unique value in the column
df['day'].unique()
#Count of unique value in the column
df['day'].nunique()
#average bill amount spent per day
df.groupby('day')['total_bill'].mean()
#average tip amount given per day
df.groupby('day')['tip'].mean()
#visualize average bill amount spent per day
avg_bill_per_day = df.groupby('day')['total_bill'].mean().reset_index()
sns.barplot(x='day', y='total_bill', data=avg_bill_per_day)
plt.title('Average Bill Amount per Day')
plt.show()
#visualize average tip amount given per day
avg_tip_per_day = df.groupby('day')['tip'].mean().reset_index()
sns.barplot(x='day', y='tip', data=avg_tip_per_day)
plt.title('Average Tip Amount per Day')
plt.show()
```
