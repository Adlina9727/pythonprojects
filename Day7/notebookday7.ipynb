```python
import pandas as pd
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Math': [85, 92, 78, 90],
    'Science': [88, 79, 95, 91],
    'English': [90, 85, 88, 92]
}

df = pd.DataFrame(data)
print(df)
```

```python
import pandas as pd
thisdict = {"student": ["Alice","Charlie","David","Eva"], "courses": ["Math","English","Math","Science"],"marks":[85,78,99,98]}
myvar = pd.DataFrame(thisdict)
print("With column names:")
print(myvar)
```

```python
import json 
thisdict = '{"student": ["Alice","Charlie","David","Eva"], "courses": ["Math","English","Math","Science"],"marks":[85,78,99,98]}'
data = json.loads(thisdict)
df = pd.DataFrame(data)
print(df)
```
