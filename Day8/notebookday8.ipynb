```python
#pretty print bar chart of student grades
#create a simple bar chart using matplotlib
import matplotlib.pyplot as plt
plt.bar(students, grades, color=['blue', 'red', 'green', 'orange'])
plt.title('Student Grades')
plt.xlabel('Students')
plt.ylabel('Grades')
plt.ylim(0, 100)
plt.show()  
```

```python
#create line graph of student grades over time
import matplotlib.pyplot as plt 
plt.plot(students, grades, marker='o', linestyle='-', color='purple')
plt.title('Student Grades Over Time')
plt.xlabel('Students')
plt.ylabel('Grades')
plt.ylim(0, 100)
plt.grid(True)
```

```python
students = ['Alice', 'Bob', 'Charlie', 'David']
grades = [85, 22, 70, 50]
```

```python
#create a pie chart of student grades
import matplotlib.pyplot as plt
labels = students
sizes = grades
colors = ['gold', 'lightcoral', 'lightskyblue', 'lightgreen']
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Student Grades Distribution')
plt.axis('equal')
plt.show()
```

```python
#All of them in the same place 
#create a figure with 1 row and 3 columns of subplots
fig, axs = plt.subplots(1, 3, figsize=(18, 5))
# Bar Chart
axs[0].bar(students, grades, color=['blue', 'red', 'green', 'orange'])
axs[0].set_title('Student Grades - Bar Chart')
axs[0].set_xlabel('Students')
axs[0].set_ylabel('Grades')
axs[0].set_ylim(0, 100)
# Line Graph
axs[1].plot(students, grades, marker='o', linestyle='-', color='purple')
axs[1].set_title('Student Grades Over Time - Line Graph')
axs[1].set_xlabel('Students')
axs[1].set_ylabel('Grades')
axs[1].set_ylim(0, 100)
axs[1].grid(True)
# Pie Chart
axs[2].pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
axs[2].set_title('Student Grades Distribution - Pie Chart') 
axs[2].axis('equal')
plt.tight_layout()
plt.show()

```
